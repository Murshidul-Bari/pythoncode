```python
# Install required packages (run this first in Google Colab)
!pip install numpy pandas matplotlib seaborn

# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import drive
import warnings
warnings.filterwarnings('ignore')
gym = pd.read_csv('/content/drive/MyDrive/GYM_data/gym_membership.csv')
```

```python
# Gym data
print("Gym data:\n",gym)
# Display last 5 rows
print("\nThe last 5 rows of the data:\n",gym.tail())
# Check data types
print("\nThe data types of the dataset:\n",gym.dtypes)
```

```python
# Task 1: Basic Matplotlib (3 plots)
# 1. Bar chart of gym membership types
# Use: gym['abonoment_type'].value_counts()

member_ship= gym ['abonoment_type'].value_counts()
plt.rcParams.update({'font.family': 'serif'})
print("\nThe Bar-Chart of gym members according to their Membership:\n")
plt.bar(['Standard','Premium'],member_ship,color=['#bc5090','#58508d'])
plt.title('Gym Membership Type')
plt.grid(True,which="both", linestyle='dotted',linewidth=1, alpha=0.7)
#plt.savefig("/content/drive/MyDrive/GYM_data/membershiptype.png", dpi=300, bbox_inches='tight')
plt.show()

# 2. Histogram of time in gym
# Use: plt.hist(gym['avg_time_in_gym'])

spendTime = gym['avg_time_in_gym']
print("\nThe histogram of gym members according to their average time in gym:\n")
plt.hist(spendTime,bins=30, color=['#5D688A'])
plt.title('Average Time Distribution of Gym Members')
plt.xlabel('Number of Gym Members')
plt.ylabel('Time Spend in Gym')
plt.grid(True, which="both", linestyle='dotted',linewidth=1, alpha=0.7)
#plt.savefig("/content/drive/MyDrive/GYM_data/avg_time_gym.png", dpi=300, bbox_inches='tight')
plt.show()

# 3. Scatter plot: Age vs visits per week
# Use: plt.scatter(gym['Age'], gym['visit_per_week'])

print("\nThe scatter plot of gym members according to their Age vs Visits per week:\n")
plt.scatter(gym['Age'], gym['visit_per_week'], alpha=0.7)
plt.title('Age vs Visits per week')
plt.xlabel('Age of the members')
plt.ylabel('Visit per week')
plt.grid(True, which="both", linestyle='-',linewidth=1, alpha=0.7)
#plt.savefig("/content/drive/MyDrive/GYM_data/visitperWeek_gym.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Task 2: Basic Seaborn (3 plots)
# 1. Box plot: Gender vs Age
# Use: sns.boxplot(data=gym, x='gender', y='Age')

sns.boxplot(data=gym, x='gender', y='Age',hue='gender')
plt.title('Overview of gym member according to gender and age:\n')
plt.xlabel("Gender")
plt.grid(True, which="both", linestyle='-',linewidth=1, alpha=0.7)
#plt.savefig("/content/drive/MyDrive/GYM_data/gender_Age.png", dpi=300, bbox_inches='tight')
plt.show()



# 2. Count plot: Days per week
# Use: sns.countplot(data=gym, x='days_per_week')
plt.figure(figsize=(25, 6))
sns.countplot(data=gym, x='days_per_week')

plt.title('Overview of gym members according to visits:\n')
plt.xticks(rotation=90)
plt.xlabel("Days per Week")
plt.ylabel("Member Count")
plt.grid(True, which="both", linestyle='-',linewidth=1, alpha=0.7)
#plt.savefig("/content/drive/MyDrive/GYM_data/day_WEEK.png", dpi=300, bbox_inches='tight')
plt.show()

# 3. Bar plot: Average visits by membership type
# Use: sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')

sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')
plt.title('Average visits by membership type:\n')
plt.xlabel('Membership Type')
plt.grid(True, which='both', linestyle='-', linewidth=0.7, alpha=0.7)
#plt.savefig("/content/drive/MyDrive/GYM_data/avg_Visits.png", dpi=300, bbox_inches='tight')
plt.show()
```
