```python
#Task 1.2: NumPy Array Operations
import numpy as np
import pandas as pd
# Convert score columns to numpy arrays
math_scores = df['math score'].values
reading_scores = df['reading score'].values
writing_scores = df['writing score'].values

# Your tasks:
# 1. Calculate mean, median, std, min, max for each subject using NumPy
# 2. Find total students who scored above 80 in math
# 3. Find total students who scored below 50 in any subject
# 4. Calculate the overall average score across all three subjects

# Expected output example:
# Math - Mean: 66.1, Median: 66.0, Std: 15.2
# Students with math > 80: 132
# Students with any score < 50: 178

# Your code here:
#Task no:1 = Calculation of Mean, Median, Std, Min, Max for each subject using NumPy
print("For Subject-Math:")
print("Mean:",round(np.mean(math_scores),1,),end=" ")
print("Median:",round(np.median(math_scores),1),end=" ")
print("Std:",round(np.std(math_scores),1))
print("Minimum Value is:",np.min(math_scores))
print("Maximum value is:",np.max(math_scores),"\n")

print("For Reading Scores:")
print("Mean:",round(np.mean(reading_scores),1),end=" ")
print("Median:",round(np.median(reading_scores),1),end=" ")
print("Std:",round(np.std(reading_scores),1))
print("Minimum Value is:",np.min(reading_scores))
print("Maximum value is:",np.max(reading_scores),"\n")

print("For Writing Scores:")
print("Mean:",round(np.mean(writing_scores),1),end=" ")
print("Median:",round(np.median(writing_scores),1),end=" ")
print("Std:",round(np.std(writing_scores),1))
print("Minimum Value is:",np.min(writing_scores))
print("Maximum value is:",np.max(writing_scores),"\n")

#Task no:2 = The total number of students who scored above 80 in math
A_grade= (math_scores)>80
print("Total number of students who scored above 80 in math:",np.sum(A_grade),"Students\n")

#Task no:3 = The total number of students who scored below 50 in any subject
allScore= df[['math score', 'reading score', 'writing score']]
badMark= (allScore<50). any(axis=1)
print("Total number of students who scored below 50 in any subject:",np.sum(badMark),"Students\n")

#Task no:4 = Calculation of the overall average score across all three subjects
print("The overall average score across all three subjects: ",round(np.mean(allScore),1))
```

```python
#Task 2.2: Create and Fix Missing Value
# Artificially create missing data for practice
import numpy as np
import pandas as pd

# Your tasks:
# 1. Make a copy of the dataset: df_practice = df.copy()
# 2. Randomly set 50 math scores to np.nan
# 3. Randomly set 30 writing scores to np.nan
# 4. Fill missing math scores with the MEDIAN math score
# 5. Fill missing writing scores with the MEAN writing score
# 6. Verify no missing values remain

# Steps to follow:
# Step 1: Create copy and show original missing values
print("Original Dataset:\n",df.isnull().sum(),"\n")


# Continue the task...
# Your code here:
# Step 2: Create missing data
#np.random.seed(42)  # For reproducible results

df_practice = df.copy()
print("Original missing values:\n", df_practice.isnull().sum())
missing_math = np.random.choice(df_practice.index, 50, replace=False)

df_practice.loc[missing_math, "math score"] = np.nan
#print(df_practice)
# Step 3: Create missing data for writing score

missing_write = np.random.choice(df_practice.index, 30, replace=False)

df_practice.loc[missing_write, "writing score"] = np.nan

#Step 4: Filling missing math score:

df_practice['math score']= df_practice['math score'].fillna(df_practice['math score'].median())
print(df_practice)

#Step 5: Filling missing write score:

df_practice['writing score']= df_practice['writing score'].fillna(df_practice['writing score'].mean())
print(df_practice)

#Step 6: Verification of no missing values

print("Remaining missing math score:", df_practice['math score'].isnull().sum())
print("Remaining missing writing score:", df_practice['writing score'].isnull().sum())
```
